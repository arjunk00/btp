```python
df=df.replace(cleanup_nums)
df
```

```python
cols_to_average = df.columns[-8:]
```

```python
df_grouped = df.groupby('speech_no')[cols_to_average].mean()
```

```python
df_grouped.sort_values('speech_no')
df_grouped
```

```python
cols_to_divide = df_grouped.columns[-8:]
```

```python
df_grouped[cols_to_divide] = df_grouped[cols_to_divide] / 10
```

```python
records=df_grouped.to_dict(orient='records')
```

```python
records
```

```python
emotion1=[]
emotion2=[]
for record in records:
    top_two_emotions = sorted(record.items(), key=lambda x: x[1], reverse=True)[:2]
    emotion1.append(top_two_emotions[0][0])
    emotion2.append(top_two_emotions[1][0])
```

```python
final_df.insert(0, "emotion1", emotion1, True)
final_df.insert(1, "emotion2", emotion2, True)
```
